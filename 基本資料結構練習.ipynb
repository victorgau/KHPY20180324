* set (集合)

### list 跟 tuple

list 跟 tuple 的主要差別是：tuple 是 immutable，意思是 tuple 的值被指定了之後就不可以改變。

譬如說：

```python
a = (1, 2, 3)
a[0] = 10
```
會出現錯誤。

### dict 跟 set

雖然 dict 跟 set 都是使用大括號 "{ }" 來表示，但是兩個資料結構是非常不一樣的。

dict 裡面存放的是鰎值對 (Key-Value Pair)，而 set 裡面存放的是不重複的元素。

另外，"{}" 表示的是空字典，空集合的表示方式是 "set()" 。

### set 的運算

* 交集：&
* 聯集：|
* 差集：-
