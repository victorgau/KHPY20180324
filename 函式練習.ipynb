# 函式練習

## 複習一下

### 函式定義：

* 無引數、無回傳值

```python
def FunctionName():
    欲執行程式區塊
```

* 有引數、無回傳值

```python
def FunctionName(arg1, arg2):
    欲執行程式區塊
```

* 兩個引數、有回傳值

```python
def FunctionName(arg1, arg2):
    欲執行程式區塊
    return 回傳值
```

### 函式呼叫：

* 無傳入引數

```python
FunctionName()
```

* 傳入兩個引數

```python
FunctionName(arg1, arg2)
```

* 傳入兩個引數，用變數承接回傳值

```python
變數 = FunctionName(arg1, arg2)
```

### 匿名函式：

```python
lambda arg1, arg2: 回傳值
```
